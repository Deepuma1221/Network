```python
from Crypto.Cipher import AES
from Crypto.Random import get_random_bytes
from Crypto.Util.Padding import pad,unpad
import os
```

```python
def encrypt_image(input_file, output_file, key):
    # Read the image file
    with open(input_file, 'rb') as file:
        image_data = file.read()

    # Create a new AES cipher
    cipher = AES.new(key, AES.MODE_ECB)

    # Pad the image data to match the AES block size
    padded_data = pad(image_data, AES.block_size)

    # Encrypt the padded data
    encrypted_data = cipher.encrypt(padded_data)

    # Write the encrypted data to the output file
    with open(output_file, 'wb') as file:
        file.write(encrypted_data)
```

```python
def decrypt_image(input_file, output_file, key):
    # Read the encrypted image file
    with open(input_file, 'rb') as file:
        encrypted_data = file.read()

    # Create a new AES cipher
    cipher = AES.new(key, AES.MODE_ECB)

    # Decrypt the encrypted data
    decrypted_data = unpad(cipher.decrypt(encrypted_data), AES.block_size)

    # Write the decrypted data to the output file
    with open(output_file, 'wb') as file:
        file.write(decrypted_data)
```

```python
# Define the input and output file paths
input_file = '/content/drive/MyDrive/ADT/1647337863145.avif'
output_directory = '/content/drive/MyDrive/ADT/encrypted/'
encrypted_file = os.path.join(output_directory, 'encrypted_image.jpg')
decrypted_file = os.path.join(output_directory, 'decrypted_image.jpg')
```

```python
# Generate a 256-bit AES key
key = get_random_bytes(32)

# Encrypt the image
encrypt_image(input_file, encrypted_file, key)

# Save the key value
key_value = key.hex()

# Print the key value
print("AES Key:", key_value)

# Decrypt the image
decrypt_image(encrypted_file, decrypted_file, key)
```
